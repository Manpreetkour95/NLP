```python
def replace_values_with_rules(input_list):
    # Initialize the output list
    output_list = []

    # Initialize a flag to track if we've encountered the first 2
    first_2_encountered = False
    b = input_list.index(':')
    # Iterate through the input list
    for i,item in enumerate(input_list):

        # If the item is '2', replace it with 0 and set the flag to True
        #print(i,item,b)
        if i==0:
            output_list.append(0)
        elif i==b+1 and item != ':':
            output_list.append(3)

        elif item==':':
            output_list.append(2)
        elif i>0 and item!=':' and i<b:
            output_list.append(1)



        else:
            output_list.append(4)

    return output_list

# Example input list
input_list = ['your', 'Phone', 'hh','Number', ':', '01', '123-456-7890','guyg']

# Call the function to replace values based on the rules
output_list = replace_values_with_rules(input_list)

# Print the resulting list
print(output_list)
```
